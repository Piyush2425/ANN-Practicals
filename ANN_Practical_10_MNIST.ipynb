# Experiment 10: MNIST Handwritten Digit Classification using TensorFlow

## Aim
To implement handwritten digit recognition using MNIST dataset.

## Objective
To understand image classification using neural networks on real-world dataset.

## Theory

### What is MNIST?

**MNIST** = Modified National Institute of Standards and Technology

A classic dataset in machine learning containing:
- **60,000 training images** (used to train the model)
- **10,000 testing images** (used to evaluate performance)
- **10 classes** (digits 0-9)
- **28×28 pixels** per image (784 features when flattened)
- **Grayscale values** (0-255, where 0=black, 255=white)

**Why MNIST?**
- Benchmark dataset for testing ML algorithms
- Simple enough to understand, complex enough to be interesting
- Real handwritten digits (not synthetic)
- Widely used for educational purposes

### Image Classification Problem

**Task:** Given a 28×28 pixel image of a handwritten digit, predict which digit (0-9) it represents.

**Challenges:**
- Different handwriting styles
- Variations in position and size
- Rotation and skewing
- Noise and partial visibility

### Neural Network Architecture

**Layer-by-Layer Processing:**

1. **Input Layer (784 neurons)**
   - 28×28 = 784 pixel values
   - Each pixel is one input feature

2. **Flatten Layer**
   - Converts 2D image (28×28) to 1D vector (784,)
   - Essential for passing image to dense layers

3. **Hidden Dense Layer (128 neurons)**
   - ReLU activation learns patterns
   - Extracts features from pixels
   - Example features: edges, corners, curves

4. **Output Layer (10 neurons)**
   - One neuron per digit (0-9)
   - Softmax activation outputs probabilities
   - Sum of all outputs = 1.0

### Key Definitions

**Flatten:**
- Converts multi-dimensional input (28, 28) to 1D array (784,)
- Necessary because Dense layers expect 1D input

**ReLU (Rectified Linear Unit):**
- Activation function: f(x) = max(0, x)
- Introduces non-linearity
- Solves vanishing gradient problem
- Most popular hidden layer activation

**Softmax:**
- Converts output layer to probability distribution
- Formula: σ(z)ᵢ = e^zᵢ / Σ e^zⱼ
- Output: 10 probabilities that sum to 1
- Example: [0.02, 0.01, 0.85, 0.01, ...]  → Predict digit 2

**Sparse Categorical Crossentropy:**
- Loss function for multi-class classification
- Used when labels are integers (0, 1, 2, ..., 9)
- Alternative: "categorical_crossentropy" with one-hot encoding

**Model Architecture Diagram:**
```
28×28 Image (784 pixels)
    ↓
Flatten Layer (784 neurons)
    ↓
Dense Layer (128 neurons, ReLU) → learns features
    ↓
Dense Layer (10 neurons, Softmax) → probability for each digit
    ↓
Output: [0.01, 0.02, 0.85, 0.01, ...] → Predict digit 2
```


````python

import tensorflow as tf
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
````

## Step 1: Import Libraries

Essential libraries for MNIST image classification:

- **TensorFlow/Keras:** Neural network framework
- **MNIST Dataset:** Pre-loaded 60,000 training + 10,000 test images
- **Matplotlib:** Visualization for images


````python

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0

print("Training data shape:", X_train.shape)
````

## Step 2: Load and Preprocess MNIST Dataset

**Dataset Loading:**
- Automatically downloads and loads MNIST from TensorFlow
- Splits into training (60,000) and testing (10,000) sets
- Returns: images and their corresponding labels (digits 0-9)

**Normalization:**
- Original pixel values: 0-255 (8-bit images)
- Normalized to: 0-1 (divide by 255)
- **Why normalize?**
  - Neural networks learn faster with normalized inputs
  - Prevents numerical instability
  - Standardizes input range
  - Recommended practice in deep learning

**Data Shape:**
- X_train: (60000, 28, 28) → 60,000 images, 28×28 pixels each
- y_train: (60000,) → 60,000 labels (0-9)


````python
import plotly.subplots as sp
import plotly.graph_objects as go
import numpy as np

# Select 10 random digits (one of each class 0-9)
sample_indices = []
for digit in range(10):
    idx = np.where(y_train == digit)[0][0]
    sample_indices.append(idx)

# Create subplot figure with 2 rows and 5 columns
fig = sp.make_subplots(
    rows=2, cols=5,
    subplot_titles=[f"Digit {i}" for i in range(10)],
    specs=[[{'type': 'image'} for _ in range(5)] for _ in range(2)]
)

# Add each digit image
for idx, digit in enumerate(range(10)):
    row = idx // 5 + 1
    col = idx % 5 + 1
    
    img = X_train[sample_indices[digit]]
    
    fig.add_trace(
        go.Image(z=img, colorscale='Gray', showscale=False, hovertemplate='Pixel value: %{z}'),
        row=row, col=col
    )

fig.update_layout(
    title_text="Sample MNIST Digits (0-9)<br><sub>Each digit shows a real handwritten example</sub>",
    title_font_size=16,
    height=500,
    width=1000,
    showlegend=False
)

fig.update_xaxes(showticklabels=False)
fig.update_yaxes(showticklabels=False)

fig.show()

print("✓ Displayed 10 sample MNIST digits (one of each class 0-9)")
print(f"  Each image is 28×28 pixels = {28*28} features")
print(f"  Pixel values range: 0 (white) to 255 (black)")
````

````python

model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])
````

## Step 3: Build Neural Network Model

**Three-Layer Architecture:**

1. **Flatten Layer**
   - Input: 28×28 image → Output: 784-dimensional vector
   - Converts spatial data to 1D for dense layers

2. **Hidden Dense Layer (128 neurons)**
   - ReLU activation: f(x) = max(0, x)
   - Learns features from the 784 inputs
   - 128 is a common choice (balance between capacity and efficiency)
   - Extracts patterns like edges, curves, corners

3. **Output Dense Layer (10 neurons)**
   - One neuron per digit class (0-9)
   - Softmax activation converts to probabilities
   - Output: 10 probability values summing to 1

**Why This Architecture?**
- Simple and effective for MNIST
- Flatten converts 2D image to 1D
- Single hidden layer sufficient for MNIST
- Softmax gives probability interpretation


````python

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
````

## Step 4: Compile Model

Prepare model for training with three key components:

**1. Optimizer: 'adam'**
- Adaptive learning rate optimizer
- Automatically adjusts learning rates per parameter
- Fast convergence, generally outperforms SGD

**2. Loss: 'sparse_categorical_crossentropy'**
- Multi-class classification loss
- "Sparse" means labels are integers (0, 1, 2, ..., 9)
- Alternative: "categorical_crossentropy" with one-hot encoded labels
- Measures how wrong predictions are

**3. Metrics: ['accuracy']**
- Monitors % of correct predictions
- Helps track training progress
- Accuracy = (correct predictions / total) × 100%

**Example:**
- True label: 3
- Prediction: [0.01, 0.02, 0.05, 0.85, 0.07] (softmax outputs)
- Predicted digit: 3 (highest probability)
- Result: ✓ Correct (accuracy += 1)


````python

history = model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
````

## Step 5: Train Model

Train the neural network on 60,000 MNIST training images.

**Training Parameters:**

1. **X_train, y_train:** Training data
   - 60,000 images + their labels
   
2. **epochs=5:** Train for 5 complete passes over dataset
   - Each epoch: 60,000 images processed once
   - More epochs = more learning but also more time
   - For MNIST: 5 epochs usually sufficient for ~95% accuracy

3. **validation_data=(X_test, y_test):** Validate on test set
   - Monitors performance on unseen data
   - Detects overfitting
   - Real measure of model generalization

**Training Process (per epoch):**
```
Forward Pass:    60,000 images → predictions
Loss Calculation: compare predictions to true labels
Backward Pass:    compute gradients (backpropagation)
Weight Update:    Adam optimizer adjusts all weights
Validation:       evaluate on test set
```

**Expected Output:**
- Loss decreases over epochs
- Accuracy increases over epochs
- Validation accuracy should track training accuracy


````python

loss, acc = model.evaluate(X_test, y_test)

print("Test Accuracy:", acc)
````

## Step 6: Evaluate Model on Test Set

Test model performance on 10,000 unseen test images.

**Evaluation Metrics:**

- **Loss:** Error on test set
  - Ideally: < 0.1 (very low)
  - > 0.5: model not learning well

- **Test Accuracy:** % correct predictions on test set
  - MNIST benchmark: 95%+ is good
  - Top models: 99.7%+
  - Our model: should achieve 97-98%

**Why Use Test Set?**
- Training set used to train weights
- Test set evaluates real-world performance
- Shows if model generalizes or just memorized
- Essential for assessing production readiness


````python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create subplots for loss and accuracy
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Training Loss", "Training Accuracy"),
    specs=[[{'type': 'scatter'}, {'type': 'scatter'}]]
)

# Add loss traces
fig.add_trace(
    go.Scatter(y=history.history['loss'], mode='lines+markers', 
               name='Training Loss', line=dict(color='red', width=2),
               marker=dict(size=8)),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(y=history.history['val_loss'], mode='lines+markers', 
               name='Validation Loss', line=dict(color='orange', width=2),
               marker=dict(size=8)),
    row=1, col=1
)

# Add accuracy traces
fig.add_trace(
    go.Scatter(y=history.history['accuracy'], mode='lines+markers', 
               name='Training Accuracy', line=dict(color='blue', width=2),
               marker=dict(size=8)),
    row=1, col=2
)
fig.add_trace(
    go.Scatter(y=history.history['val_accuracy'], mode='lines+markers', 
               name='Validation Accuracy', line=dict(color='green', width=2),
               marker=dict(size=8)),
    row=1, col=2
)

# Update axes
fig.update_xaxes(title_text="Epoch", row=1, col=1)
fig.update_xaxes(title_text="Epoch", row=1, col=2)
fig.update_yaxes(title_text="Loss", row=1, col=1)
fig.update_yaxes(title_text="Accuracy", row=1, col=2)

fig.update_layout(
    title_text="MNIST Model Training Progress<br><sub>Loss and Accuracy over 5 epochs</sub>",
    title_font_size=16,
    hovermode='x unified',
    height=500,
    width=1200
)

fig.show()

print("✓ Training Complete!")
print(f"  Final Training Accuracy: {history.history['accuracy'][-1]:.2%}")
print(f"  Final Validation Accuracy: {history.history['val_accuracy'][-1]:.2%}")
````

````python

predictions = model.predict(X_test[:5])

for i in range(5):
    plt.imshow(X_test[i], cmap='gray')
    plt.title(f"Predicted: {predictions[i].argmax()}")
    plt.show()
````

## Step 7: Make Predictions on Test Images

Visualize model predictions on first 5 test images with probabilities.

**Process:**
1. Get predictions from model
2. For each image:
   - Display the 28×28 pixel image
   - Show predicted digit
   - Show confidence (probability)

**Interpretation:**
- Prediction = argmax of output layer
- Example output: [0.01, 0.02, 0.85, 0.01, 0.05, 0.02, 0.01, 0.01, 0.01, 0.01]
- Prediction: digit 2 (index with highest probability 0.85)
- Confidence: 85%


````python
import plotly.subplots as sp

# Get predictions for first 10 test images
predictions = model.predict(X_test[:10])
predicted_digits = predictions.argmax(axis=1)

# Create visualization with subplots (2 rows x 5 cols)
fig = sp.make_subplots(
    rows=2, cols=5,
    subplot_titles=[f"True: {y_test[i]}, Pred: {predicted_digits[i]}<br>Conf: {predictions[i].max():.1%}" 
                    for i in range(10)],
    specs=[[{'type': 'image'} for _ in range(5)] for _ in range(2)]
)

# Add images
for idx in range(10):
    row = idx // 5 + 1
    col = idx % 5 + 1
    
    img = X_test[idx]
    
    fig.add_trace(
        go.Image(z=img, colorscale='Gray', showscale=False,
                hovertemplate='Pixel value: %{z}'),
        row=row, col=col
    )

fig.update_layout(
    title_text="MNIST Predictions on Test Images<br><sub>Shows True Label, Predicted Label, and Confidence</sub>",
    title_font_size=16,
    height=500,
    width=1000,
    showlegend=False
)

fig.update_xaxes(showticklabels=False)
fig.update_yaxes(showticklabels=False)

fig.show()

print("\n✓ Predictions Summary:")
for i in range(10):
    true_label = y_test[i]
    pred_label = predicted_digits[i]
    confidence = predictions[i].max()
    is_correct = "✓" if true_label == pred_label else "✗"
    print(f"  Image {i}: True={true_label}, Pred={pred_label}, Confidence={confidence:.1%} {is_correct}")
````

### Confidence Distribution Visualization

Let's visualize the probability outputs from the softmax layer for the first 3 test images. This shows how confident the model is about each digit class.


````python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Get predictions for first 3 images
sample_predictions = model.predict(X_test[:3])

# Create subplots for 3 bar charts
fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=[f"Image 0 (True: {y_test[0]})" , 
                    f"Image 1 (True: {y_test[1]})",
                    f"Image 2 (True: {y_test[2]})"],
    specs=[[{'type': 'bar'}, {'type': 'bar'}, {'type': 'bar'}]]
)

# Add bar traces for each image
for img_idx in range(3):
    probabilities = sample_predictions[img_idx]
    predicted_digit = probabilities.argmax()
    
    # Color code: green for correct prediction, red for incorrect
    colors = ['green' if i == predicted_digit else 'lightgray' for i in range(10)]
    
    fig.add_trace(
        go.Bar(
            x=list(range(10)),
            y=probabilities,
            name=f'Image {img_idx}',
            marker=dict(color=colors),
            text=[f'{p:.2%}' for p in probabilities],
            textposition='outside',
            hovertemplate='Digit %{x}<br>Probability: %{y:.2%}<extra></extra>'
        ),
        row=1, col=img_idx+1
    )

# Update layout
fig.update_layout(
    title_text="Softmax Output Distribution<br><sub>Model confidence for each digit class (0-9)</sub>",
    title_font_size=16,
    height=500,
    width=1400,
    showlegend=False
)

# Update axes
for i in range(1, 4):
    fig.update_xaxes(title_text="Digit", row=1, col=i)
    fig.update_yaxes(title_text="Probability", row=1, col=i)

fig.show()

print("\n✓ Softmax Output Analysis:")
print("  Green bar = predicted digit")
print("  Height = confidence (probability)")
print("  Sum of all probabilities = 100%")
````

4. Train for more epochs

1. Use Convolutional Neural Networks (CNNs)
2. Replace Dense layers with Conv2D layers
3. Add pooling layers for dimensionality reduction
4. Typically achieves 99%+ accuracy
