
# Experiment 4: Perceptron Learning Law with Decision Regions

## Aim
To demonstrate the perceptron learning law and visualize decision regions.

## Objective
To understand how perceptron learns and separates data using a decision boundary.

## Theory
Perceptron is a supervised learning algorithm for binary classification.

### Equation:
Y = 1 if (W·X + b ≥ 0) else 0

### Learning Rule:
W = W + η(d - y)x  
b = b + η(d - y)

Where:
- d = desired output
- y = predicted output
- η = learning rate

The perceptron creates a **linear decision boundary** to separate classes.


```python

import numpy as np
import plotly.graph_objects as go

# Simple linearly separable dataset
X = np.array([
    [2, 3],
    [3, 4],
    [5, 6],
    [6, 7],
    [1, 1],
    [2, 1]
])

# Labels: 1 and 0
y = np.array([1, 1, 1, 1, 0, 0])
```

```python

# Initialize weights and bias
w = np.random.randn(2)
b = np.random.randn()

learning_rate = 0.1
epochs = 50
```

```python

# Training perceptron
for epoch in range(epochs):
    for i in range(len(X)):
        linear = np.dot(X[i], w) + b
        y_pred = 1 if linear >= 0 else 0

        # Update rule
        w += learning_rate * (y[i] - y_pred) * X[i]
        b += learning_rate * (y[i] - y_pred)

print("Weights:", w)
print("Bias:", b)
```

```python

# Decision boundary: w1*x1 + w2*x2 + b = 0
# => x2 = -(w1*x1 + b) / w2

x_vals = np.linspace(0, 7, 100)
y_vals = -(w[0]*x_vals + b) / w[1]
```

```python

# Plot data points
fig = go.Figure()

# Class 1
fig.add_trace(go.Scatter(
    x=X[y==1][:,0], y=X[y==1][:,1],
    mode='markers',
    name='Class 1'
))

# Class 0
fig.add_trace(go.Scatter(
    x=X[y==0][:,0], y=X[y==0][:,1],
    mode='markers',
    name='Class 0'
))

# Decision boundary
fig.add_trace(go.Scatter(
    x=x_vals, y=y_vals,
    mode='lines',
    name='Decision Boundary'
))

fig.update_layout(
    title="Perceptron Decision Region",
    xaxis_title="X1",
    yaxis_title="X2"
)

fig.show()
```


## Conclusion
We successfully implemented perceptron learning and visualized the decision boundary separating two classes.
