# Experiment 1: Activation Functions

## Aim
To plot and understand different activation functions used in neural networks.

## Objective
To learn how activation functions introduce non-linearity and visualize them using Python.

## Theory
Activation functions introduce non-linearity into neural networks.

### Functions:
- Linear: f(x) = x
- Sigmoid: f(x) = 1 / (1 + e^(-x))
- Tanh: f(x) = tanh(x)
- ReLU: f(x) = max(0, x)
- Softmax: probability distribution

They help neural networks learn complex patterns.


```python

import numpy as np
import plotly.graph_objects as go

# Define activation functions
def linear(x):
    return x

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()
```

```python
import numpy as np
import plotly.graph_objects as go

# Functions
def linear(x):
    return x

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Input
x = np.linspace(-10, 10, 100)

# Plot
fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=linear(x), name='Linear'))
fig.add_trace(go.Scatter(x=x, y=sigmoid(x), name='Sigmoid'))
fig.add_trace(go.Scatter(x=x, y=tanh(x), name='Tanh'))
fig.add_trace(go.Scatter(x=x, y=relu(x), name='ReLU'))

fig.update_layout(
    title="Activation Functions",
    xaxis_title="Input",
    yaxis_title="Output"
)

fig.show()
```

```python

# 3D Sigmoid Surface
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

X, Y = np.meshgrid(x, y)
Z = 1 / (1 + np.exp(-(X + Y)))

fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])

fig.update_layout(title='3D Sigmoid Surface',
                  scene=dict(
                      xaxis_title='X1',
                      yaxis_title='X2',
                      zaxis_title='Output'))

fig.show()
```

```python

# Output check
x = np.array([-1, 0, 1])

print("Sigmoid:", sigmoid(x))
print("Tanh:", tanh(x))
print("ReLU:", relu(x))
print("Softmax:", softmax(x))
```

## Mathematical Intuition of Activation Functions

Each activation function has unique properties that affect how neural networks learn:

- **Linear**: f(x) = x | No non-linearity; limited learning capacity
- **Sigmoid**: f(x) = 1/(1+e^(-x)) | Outputs between 0-1; smooth curve; prone to vanishing gradients
- **Tanh**: f(x) = tanh(x) | Outputs between -1 to 1; zero-centered; better than sigmoid
- **ReLU**: f(x) = max(0,x) | Outputs 0 or x; fast computation; allows some gradients to die
- **Softmax**: Converts outputs to probability distribution; used in multi-class classification

The **derivative** (gradient) of each function determines how well the network learns during backpropagation.

```python
# Compute derivatives for each activation function
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Create individual plots with derivatives
x = np.linspace(-10, 10, 200)

# 1. Linear Function
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=x, y=linear(x), mode='lines', name='f(x) = x', 
                          line=dict(color='blue', width=3)))
fig1.add_trace(go.Scatter(x=x, y=np.ones_like(x), mode='lines', name="f'(x) = 1 (derivative)", 
                          line=dict(color='red', width=2, dash='dash')))
fig1.update_layout(title='<b>Linear Activation</b><br>f(x) = x<br><i>Non-linearity: None | Output range: (-∞, +∞)</i>',
                   xaxis_title='Input', yaxis_title='Output', hovermode='x unified', height=450)
fig1.show()

# 2. Sigmoid Function
fig2 = go.Figure()
fig2.add_trace(go.Scatter(x=x, y=sigmoid(x), mode='lines', name='f(x) = 1/(1+e^(-x))', 
                          line=dict(color='green', width=3)))
fig2.add_trace(go.Scatter(x=x, y=sigmoid_derivative(x), mode='lines', name="f'(x) derivative", 
                          line=dict(color='red', width=2, dash='dash')))
fig2.add_hline(y=0.5, line_dash="dot", line_color="gray", annotation_text="Output = 0.5")
fig2.update_layout(title='<b>Sigmoid Activation</b><br>f(x) = 1/(1+e^(-x))<br><i>Non-linearity: Strong | Output range: (0, 1) | Issue: Vanishing gradients</i>',
                   xaxis_title='Input', yaxis_title='Output', hovermode='x unified', height=450)
fig2.show()

# 3. Tanh Function
fig3 = go.Figure()
fig3.add_trace(go.Scatter(x=x, y=tanh(x), mode='lines', name='f(x) = tanh(x)', 
                          line=dict(color='purple', width=3)))
fig3.add_trace(go.Scatter(x=x, y=tanh_derivative(x), mode='lines', name="f'(x) derivative", 
                          line=dict(color='red', width=2, dash='dash')))
fig3.add_hline(y=0, line_dash="dot", line_color="gray", annotation_text="Zero-centered")
fig3.update_layout(title='<b>Tanh (Hyperbolic Tangent) Activation</b><br>f(x) = tanh(x)<br><i>Non-linearity: Strong | Output range: (-1, 1) | Better than Sigmoid: Zero-centered</i>',
                   xaxis_title='Input', yaxis_title='Output', hovermode='x unified', height=450)
fig3.show()

# 4. ReLU Function
fig4 = go.Figure()
fig4.add_trace(go.Scatter(x=x, y=relu(x), mode='lines', name='f(x) = max(0, x)', 
                          line=dict(color='orange', width=3)))
fig4.add_trace(go.Scatter(x=x, y=relu_derivative(x), mode='lines', name="f'(x) derivative", 
                          line=dict(color='red', width=2, dash='dash')))
fig4.update_layout(title='<b>ReLU (Rectified Linear Unit) Activation</b><br>f(x) = max(0, x)<br><i>Non-linearity: Simple | Output range: [0, +∞) | Advantage: Fast, avoids vanishing gradients | Disadvantage: Dying ReLU</i>',
                   xaxis_title='Input', yaxis_title='Output', hovermode='x unified', height=450)
fig4.show()

# 5. Softmax (visualized with multiple inputs)
fig5 = go.Figure()
x_softmax = np.array([-2, -1, 0, 1, 2])
softmax_out = softmax(x_softmax)
fig5.add_trace(go.Bar(x=x_softmax, y=softmax_out, name='Softmax Output', 
                      marker=dict(color='cyan')))
fig5.update_layout(title='<b>Softmax Activation</b><br>Converts outputs to probability distribution<br><i>Use: Multi-class classification | Output: Sum = 1.0 | Range: (0, 1)</i>',
                   xaxis_title='Input Values', yaxis_title='Probability', height=450)
fig5.show()
```

```python
# Comparison: Derivatives of All Activation Functions
# This shows gradient flow during backpropagation

x = np.linspace(-10, 10, 200)

fig_derivatives = go.Figure()

# Plot derivatives
fig_derivatives.add_trace(go.Scatter(x=x, y=np.ones_like(x), name='Linear Derivative', 
                                      line=dict(color='blue', width=2)))
fig_derivatives.add_trace(go.Scatter(x=x, y=sigmoid_derivative(x), name='Sigmoid Derivative', 
                                      line=dict(color='green', width=2)))
fig_derivatives.add_trace(go.Scatter(x=x, y=tanh_derivative(x), name='Tanh Derivative', 
                                      line=dict(color='purple', width=2)))
fig_derivatives.add_trace(go.Scatter(x=x, y=relu_derivative(x), name='ReLU Derivative', 
                                      line=dict(color='orange', width=2)))

fig_derivatives.update_layout(
    title='<b>Comparison of Derivatives (Gradient Flow)</b><br><i>Higher derivative = stronger gradient signal during backpropagation</i>',
    xaxis_title='Input',
    yaxis_title='Derivative Value',
    hovermode='x unified',
    height=450,
    annotations=[
        dict(text='Sigmoid & Tanh: Small gradients at extremes<br>(Vanishing Gradient Problem)', 
             x=-5, y=0.25, showarrow=True, arrowhead=2),
        dict(text='ReLU: Constant gradient for x>0<br>(No vanishing gradient)', 
             x=5, y=0.5, showarrow=True, arrowhead=2)
    ]
)

fig_derivatives.show()

# Summary Statistics
print("\n" + "="*60)
print("ACTIVATION FUNCTIONS: MATHEMATICAL PROPERTIES")
print("="*60)

x_test = np.array([-5, -1, 0, 1, 5])
print(f"\nInput values: {x_test}\n")

print("LINEAR:")
print(f"  Output:     {linear(x_test)}")
print(f"  Derivative: {np.ones_like(x_test)}")
print(f"  Characteristics: No non-linearity, constant gradient")

print("\nSIGMOID:")
print(f"  Output:     {np.round(sigmoid(x_test), 4)}")
print(f"  Derivative: {np.round(sigmoid_derivative(x_test), 4)}")
print(f"  Characteristics: S-shaped curve, outputs in [0,1], prone to vanishing gradients")

print("\nTANH:")
print(f"  Output:     {np.round(tanh(x_test), 4)}")
print(f"  Derivative: {np.round(tanh_derivative(x_test), 4)}")
print(f"  Characteristics: Similar to sigmoid but zero-centered, better gradient flow")

print("\nRELU:")
print(f"  Output:     {relu(x_test)}")
print(f"  Derivative: {relu_derivative(x_test)}")
print(f"  Characteristics: Simple, fast, constant gradients for positive inputs")

print("\nSOFTMAX (example with 5 inputs):")
test_inputs = np.array([1, 2, 3, 4, 5])
print(f"  Input:  {test_inputs}")
print(f"  Output: {np.round(softmax(test_inputs), 4)}")
print(f"  Sum of probabilities: {np.round(softmax(test_inputs).sum(), 4)}")
print(f"  Use: Multi-class classification (outputs sum to 1)")

print("="*60)
```
