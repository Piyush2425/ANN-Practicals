
# Experiment 3: Perceptron Neural Network (Even/Odd Classification)

## Aim
To classify numbers (0–9) into even and odd using a perceptron.

## Objective
To understand perceptron learning and implement binary classification.

## Theory
A perceptron is a single-layer neural network used for binary classification.

### Formula:
Y = 1 if (W·X + b ≥ 0) else 0

Where:
- W = weight
- X = input
- b = bias

### Learning Rule:
W = W + α(y - ŷ)X  
b = b + α(y - ŷ)

Where:
- y = actual output
- ŷ = predicted output
- α = learning rate


```python

import numpy as np

# Input data (0–9)
X = np.array([[i] for i in range(10)])

# Labels: Even = 1, Odd = 0
y = np.array([1 if i % 2 == 0 else 0 for i in range(10)])

print("Inputs:", X.T)
print("Labels:", y)
```

```python

# Initialize parameters
w = np.random.randn(1)
b = np.random.randn()
learning_rate = 0.1
epochs = 50

print("Initial Weight:", w)
print("Initial Bias:", b)
```

```python

# Training perceptron
for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], w) + b
        y_pred = 1 if linear_output >= 0 else 0

        # Update rule
        w += learning_rate * (y[i] - y_pred) * X[i]
        b += learning_rate * (y[i] - y_pred)

print("Trained Weight:", w)
print("Trained Bias:", b)
```

```python

# Testing predictions
predictions = []

for i in range(len(X)):
    linear_output = np.dot(X[i], w) + b
    y_pred = 1 if linear_output >= 0 else 0
    predictions.append(y_pred)

print("Predictions:", predictions)
print("Actual     :", y.tolist())
```

```python

import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=X.flatten(), y=predictions, mode='markers',
                         marker=dict(size=10, color=predictions, colorscale='Viridis'),
                         name='Predictions'))
fig.add_trace(go.Scatter(x=X.flatten(), y=y, mode='markers',
                         marker=dict(size=8, symbol='x', color='red'),
                         name='Actual'))

fig.update_layout(
    title='Perceptron Classification: Even(1) vs Odd(0)',
    xaxis_title='Number',
    yaxis_title='Even(1)/Odd(0)',
    hovermode='x unified',
    height=450
)

fig.show()
```


## Conclusion
We successfully implemented a perceptron to classify numbers as even and odd.
